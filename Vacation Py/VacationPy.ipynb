```python
#Load the csv exported in Part I to a DataFrame

#Read CSV File
output_file = "../WeatherPy/output_data/City_Data.csv"
weather = pd.read_csv(output_file)
weather
```

```python
#Humidity Heatmap

gmaps.configure(api_key = g_key)
locations = weather[["Lat", "Lng"]]
hum = weather["Humidity"]

#Plot Gmaps
fig = gmaps.figure()

#Create a heat layer
fig.add_layer(gmaps.heatmap_layer(locations, weights = hum, max_intensity=100))
fig
```

```python
#Create new DataFrame fitting weather criteria

#Wind speed less than 5 mph.
#Cloudinss of less than 10%
#Max temp higher than 75.


ideal_weather = weather.loc[(weather["Wind Speed"]<5)&(weather["Cloudiness"]<10)&(weather["Max Temp"]>75)]
ideal_weather


#Drop any rows will null value
ideal_weather = ideal_weather.dropna(how="any")
len(ideal_weather)
```

```python
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
response_list = []

for index,row in ideal_weather.iterrows():
    lat = row["Lat"]
    lng = row["Lng"]
    locations_ideal = (f"{lat},{lng}")
    
    #add key to params
    params = {
    "type": "lodging",
    "keyword": "hotel",
    "radius": 5000,
    "key": g_key,
    "location": locations_ideal
    }

    response = requests.get(base_url, params=params).json()
    response_list.append(response)
    pprint(response)
pprint(response_list)
```

```python
len(response_list)
```

```python
hotel_name = []

for x in range(len(response_list)):
    try:
        hotel_name.append(response_list[x]["results"][0]["name"])     
    except:
        hotel_name.append("NaN")

hotel_name
```

```python
#Add Hotel Name to Ideal Weather DF

ideal_weather["Hotel Name"] = hotel_name
hotel_df["Hotel Name"] = ideal_weather["Hotel Name"].astype(str)
hotel_df
```
